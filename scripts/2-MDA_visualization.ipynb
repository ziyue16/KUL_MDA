```python
# Packages
import pandas as pd
import numpy as np
import os
import altair as alt # pip install altair
from altair_saver import save # pip install altair_saver
from pyecharts import options as opts     # pip install -i https://pypi.tuna.tsinghua.edu.cn/simple pyecharts
from pyecharts.charts import  Timeline,Map

# set output path
os.chdir("")

# importing data from S3
project_dir = ("https://s3groupaustralia.s3.eu-central-1.amazonaws.com/data/data+viz/")
```

```python
# Data import
df = pd.DataFrame(pd.read_csv(project_dir + "Predict_Heatwaves.csv")) 
month = pd.DataFrame(pd.read_excel(project_dir + "HeatWaves.xlsx")) # original sheet 2000-2020 
```

```python
# Data processing  
df['Unnamed: 0']=0
df = df.drop(['heatwave'],axis=1)
df= df.drop_duplicates() 
df.index = df['Dis No']
df['tavg_F'] = df.tavg_F.round(2)
df['HI'] = df.HI.round(2)
df['Rel_Hum'] = df.Rel_Hum.round(2)
```

```python
df
```

```python
df['Date_current']= df['Dis No']
#df['Date_current_time']= df['Dis No']
df['Heatwave_Year']= df['Dis No']

for i in range(len(df)):
    year = df['Year'][i]
    month = df['Month'][i]
    df['Date_current'][i] = str(year) + '-' + str(month)
#for i in range(len(df)):
#    df['Date_current_time'][i] = pd.to_datetime(df['Date_current'][i], format='%Y-%m')
for i in range(len(df)):
    df['Heatwave_Year'][i] = df['Dis No'][i][0:4]
```

```python
month = pd.DataFrame(pd.read_excel(project_dir + "HeatWaves.xlsx")) # original sheet 2000-2020 
```

```python
month = month[['DisNo','EndMonth','TotalDeaths']] #  Some countries heatwave last for more than 1 month, End month shall prevail
month.columns = ['Dis No','Heatwave_Month','TotalDeaths']
df = df.drop(['Dis No'],axis=1)
df_base = df.merge(month,on='Dis No',how='left')
df_base = df_base.drop_duplicates() 
```

## Chart 1

```python
# Chart1
slider = alt.binding_range(min=2000, max=2020, step=1)
selector = alt.selection_single(name="Year", fields=['Year'],
                                bind=slider)
bar = alt.Chart(df_chart1).mark_bar().encode(
    x='Month:N',
    y='Rel_Hum:Q'
)

line =  alt.Chart(df_chart1).mark_line(color='red',size = 0.5).encode(
    y='tavg:Q',
    x='Month:N'
)

c = bar + line

c = c.add_selection(
    selector
).transform_filter(
    selector
)
```

```python
# filter countries

DisNo_filter = alt.binding_select(options=dis_number)
DisNo_select = alt.selection_single(fields=['Dis No'], bind=DisNo_filter, name="dis number")

chart1 = c.add_selection(
    DisNo_select
).transform_filter(
    DisNo_select
)
alt.data_transformers.disable_max_rows()
chart1
save(chart1, "chart1(line&bar).html")
```

## Chart 2

```python
# Chart2
chart2 = alt.Chart(df_chart2).mark_bar(size = 10).encode(
    y=alt.Y('Year:N'),
    x='sum(TotalDeaths)',
    color=alt.Color('Country',legend=None),
    tooltip = ['Country','TotalDeaths','Dis No']
).interactive()

chart2 
save(chart2, "chart2(stack_bar).html")
```

## Chart 3

```python
# Chart3
chart3 = Timeline()
for i in range(2000, 2021):
        df_all_layer = df_chart3.loc[df_chart3['Year']==i]
        map0 = (
        Map()
        .add("HI", [list(z) for z in zip(df_all_layer['Countrynew'], df_all_layer['HI'])], "world",is_map_symbol_show=False)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-{}".format(i)),
            visualmap_opts=opts.VisualMapOpts(max_=150,min_=50),
        )
    )
        chart3.add(map0, "{}Year".format(i))
chart3.render("chart3(map).html")
```

## Chart 4

```python
# Chart4
# change date as time variable
alt.data_transformers.disable_max_rows()
scales = alt.selection_interval(bind='scales')
nearest = alt.selection(type='single', nearest=True, on='mouseover',fields=['Dis No'])


line = alt.Chart(df_chart4).mark_line().encode(
    alt.X('Date_current:T'),
    alt.Y('HI:Q'),
    alt.Color('Dis No:N'),
    #tooltip = ["Country","Year","Month_x","Year_HW","Month_y","signal","HI","Dis No"]
).add_selection(
    scales
)

rule = alt.Chart(df_chart4).mark_rule().encode(
    y='mean(HI)',
    color=alt.value('red'),
    size=alt.value(0.5)
)

point = alt.Chart(df_chart4).mark_point().encode(
    x = 'Date_current:T',
    y = 'HI:Q',
    size=alt.value(1),
    color=alt.value('red'),
).transform_filter(alt.datum.single == 1)

selectors = alt.Chart(df_chart4).mark_point().encode(
    x='Date_current:T',
    y='HI:Q',
    tooltip = ["Dis No","Country","Year","Month","signal","HI"],
    opacity=alt.value(0),
).add_selection(
    nearest
)

chart = line + rule + point + selectors
# filter countries
country_filter = alt.binding_select(options=country_name)
country_select = alt.selection_single(fields=['Country'], bind=country_filter, name="country")

chart4 = chart.add_selection(
    country_select
).transform_filter(
    country_select
)

save(chart4, "chart4(line).html")
```
